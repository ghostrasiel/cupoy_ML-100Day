```python
from sklearn.ensemble import RandomForestClassifier , RandomForestRegressor
from sklearn import datasets
from sklearn.metrics import mean_squared_error  , accuracy_score
from sklearn.model_selection import train_test_split , cross_val_score
```

```python
wine = datasets.load_wine()
x_train , x_test , y_train , y_test = train_test_split(wine.data , wine.target , random_state=1 , train_size=0.8)
c_tree = RandomForestClassifier(n_estimators=12 , min_samples_leaf=5 , min_samples_split=3)
c_tree.fit(x_train , y_train)
y_pred = c_tree.predict(x_test)
accuracy_score(y_test , y_pred)
```

```python
boston = datasets.load_boston()
x_train , x_test , y_train , y_test = train_test_split(boston.data , boston.target , random_state=1 , train_size=0.8)
r_tree = RandomForestRegressor(n_estimators=10)
r_tree.fit(x_train , y_train)
y_pred = r_tree.predict(x_test)
y_pred
```

```python
mean_squared_error(y_test , y_pred)
```

```python
import matplotlib.pyplot as plt
plt.plot(y_test)
plt.plot(y_pred)
plt.show()
```
