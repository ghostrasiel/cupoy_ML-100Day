```python
# 載入套件
import time
import warnings

import numpy as np
import matplotlib.pyplot as plt

from sklearn import cluster, datasets
from sklearn.preprocessing import StandardScaler
from itertools import cycle, islice
```

```python
# 設定 2D 樣板資料
n_samples = 1500
random_state = 100

# 生成 雙半月 資料點
noisy_moons = datasets.make_moons(n_samples=n_samples, noise=.05)

# 生成 緊密三群 資料點
blobs = datasets.make_blobs(n_samples=n_samples, random_state=8)

# 生成 2維均勻分布 資料點
no_structure = np.random.rand(n_samples, 2), None
```

```python
# 設定群集與繪圖參數
plt.figure(figsize=(10, 8))
plot_num = 1
default_base = {'n_neighbors': 10, 'n_clusters': 3}
datasets = [(noisy_moons, {'n_clusters': 2}), (blobs, {'n_neighbors': 10, 'n_clusters': 3}), 
            (no_structure, {'n_neighbors': 10, 'n_clusters': 3})]
```

# 作業
* 試著完成剩下三種 2D 樣板資料的繪圖, 並觀察有何不同

```python
# 執行各種樣板資料的繪圖迴圈
i = 0
linkage_list = ['ward' , 'complete' , 'average']
plt.figure(figsize=(15,14))
for datas , parm in datasets:
    X = StandardScaler().fit_transform(datas[0])
    
    for linkage in linkage_list :
        i += 1
        hc = cluster.AgglomerativeClustering(n_clusters=parm['n_clusters'] , linkage=linkage)
        y_pred = hc.fit_predict(X)

        plt.subplot(3,3,i)
        plt.scatter(X[: , 0] , X[: , 1] , c=y_pred)
        plt.title(linkage)
plt.show()
```
