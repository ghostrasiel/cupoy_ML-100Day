```python
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier , DecisionTreeRegressor
from sklearn.metrics import mean_squared_error , accuracy_score
from sklearn.model_selection import train_test_split ,cross_val_score
```

```python
wine =datasets.load_wine()
x_train , x_test , y_train , y_test = train_test_split(wine.data , wine.target ,  train_size=0.8 , random_state=1)
y_train
```

```python
clf = DecisionTreeClassifier(min_samples_leaf=5)
clf.fit(x_train , y_train)
y_pred = clf.predict(x_test)
accuracy_score(y_test , y_pred)
```

```python
cross_val_score(clf , x_train , y_train , cv=10).mean()
```

```python
#DecisionTreeRegressor
boston = datasets.load_boston()
x_train , x_test , y_train , y_test = train_test_split(boston.data , boston.target ,  train_size=0.8 , random_state=1)
y_train
```

```python
rlf = DecisionTreeRegressor()
rlf.fit(x_train , y_train)
y_pred = rlf.predict(x_test)
y_pred
```

```python
mean_squared_error(y_test , y_pred)
```
