```python
from sklearn import datasets
digits = datasets.load_digits()
len(digits.target)
```

```python
from sklearn.ensemble import GradientBoostingClassifier , GradientBoostingRegressor
from sklearn.metrics import mean_squared_error , accuracy_score , confusion_matrix , ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split 
import matplotlib.pyplot as plt
```

```python
plt.imshow(digits.images[3])
plt.show()
```

```python
x_train , x_test , y_train , y_test = train_test_split(digits.data , digits.target , random_state=1, train_size=0.8)
model = GradientBoostingClassifier(learning_rate=0.05 , n_estimators=50)
model.fit(x_train , y_train)
y_pred = model.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(y_pred=y_pred , y_true=y_test)  , display_labels=digits.target_names).plot()
```

```python
boston = datasets.load_boston()
x_train , x_test , y_train , y_test = train_test_split(boston.data , boston.target , random_state=1, train_size=0.8)
model_r = GradientBoostingRegressor(learning_rate=0.05)
model_r.fit(x_train , y_train)
y_pred = model_r.predict(x_test)
y_pred
```

```python
mean_squared_error(y_test , y_pred)
```

```python
plt.plot(y_test)
plt.plot(y_pred , alpha=0.7)
plt.show()
```
