# 作業
* 試著模擬出 5 群高斯分布的資料, 並以此觀察 K-mean 與輪廓分析的結果  

```python
# 載入套件
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn import datasets
from sklearn.metrics import silhouette_samples, silhouette_score

np.random.seed(5)

%matplotlib inline
```

```python
# 生成 5 群資料
X, y = make_blobs(n_samples=500,
                  n_features=2,
                  centers=5,
                  cluster_std=1,
                  center_box=(-10.0, 10.0),
                  shuffle=True,
                  random_state=123) 

# 設定需要計算的 K 值集合
range_n_clusters = [2, 3, 4, 5, 6, 7, 8]
```

```python
# 計算並繪製輪廓分析的結果
kmeans_list = []
for k in range_n_clusters :
    km = KMeans(n_clusters=k )
    km.fit(X)
    score = silhouette_score(X ,km.labels_)
    kmeans_list.append(score)
print(kmeans_list)
    
```

```python
from kneed import KneeLocator #拐點運算
si_kl = KneeLocator(range_n_clusters , kmeans_list , curve="concave", direction="increasing")
print(si_kl.elbow , si_kl.all_elbows_y)
```

```python
plt.plot(range(2,9),kmeans_list , '-o' , zorder=20)
plt.scatter(si_kl.elbow , si_kl.all_elbows_y , c='r' , zorder=30)
```

```python
model  = KMeans(n_clusters=3)
model.fit(X)
model.inertia_
```

```python
c = model.cluster_centers_
plt.scatter(X[: , 0] , X[: , 1] , c=model.labels_)
plt.scatter(c[:,0] , c[:,1] , c='red' , alpha=0.5 , s=200)
```
