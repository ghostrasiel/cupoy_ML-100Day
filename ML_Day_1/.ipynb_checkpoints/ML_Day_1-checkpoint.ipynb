```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w = 3
b = 0.5
x_lin = np.linspace(0 , 100 , 100) #1 到 100 產生 100個點
x_lin
```

```python
y = (x_lin + np.random.randn(100) * 5) * w + b
plt.plot(x_lin   , y , 'b.' , label='points')
plt.legend(loc=2)
plt.show()
```

```python
y_hat = x_lin *w +b
plt.plot(x_lin , y , 'b.' , label='data')
plt.plot(x_lin , y_hat ,'r-' ,label='prediction')
plt.legend(loc=2)
plt.show()
```
